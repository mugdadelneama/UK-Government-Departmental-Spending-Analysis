# Checking which categories or suppliers increased or decreased 

```python
# Group by year and expense type
df['Date of Payment'] = pd.to_datetime(df['Date of Payment'], dayfirst=True)
yearly_category = df.groupby([df['Date of Payment'].dt.year, 'Expense Type'])['Amount'].sum().reset_index()
yearly_category.columns = ['Year', 'Expense Type', 'Total Spend']

# Split into 2024 and 2025
spend_2024 = yearly_category[yearly_category['Year'] == 2024].set_index('Expense Type')['Total Spend']
spend_2025 = yearly_category[yearly_category['Year'] == 2025].set_index('Expense Type')['Total Spend']

# Compare and calculate change
comparison = pd.DataFrame({'2024': spend_2024, '2025': spend_2025}).fillna(0)
comparison['Change'] = comparison['2025'] - comparison['2024']
comparison['Change %'] = ((comparison['Change'] / comparison['2024']) * 100).round(2)
comparison = comparison.sort_values('Change', ascending=False)

print("ðŸ“Š Category Spend Change (2024 vs 2025):")
print(comparison)

# Show increased and decreased categories
print("\nðŸ“ˆ Increased Categories:")
print(comparison[comparison['Change'] > 0][['2024', '2025', 'Change %']])

print("\nðŸ“‰ Decreased Categories:")
print(comparison[comparison['Change'] < 0][['2024', '2025', 'Change %']])
```

# What changed between years ?

```python
# Convert date and extract year
df['Date of Payment'] = pd.to_datetime(df['Date of Payment'])
df['Year'] = df['Date of Payment'].dt.year

# Compare categories between 2024 and 2025
category_2024 = df[df['Year'] == 2024].groupby('Expense Type')['Amount'].sum()
category_2025 = df[df['Year'] == 2025].groupby('Expense Type')['Amount'].sum()

comparison = pd.DataFrame({'2024': category_2024, '2025': category_2025}).fillna(0)
comparison['Change Â£'] = comparison['2025'] - comparison['2024']
comparison['Change %'] = ((comparison['Change Â£'] / comparison['2024']) * 100).round(2)
comparison = comparison.sort_values('Change Â£', ascending=False)

print(comparison)
```

```python
# Step 1: Make sure date is datetime
df['Date of Payment'] = pd.to_datetime(df['Date of Payment'], dayfirst=True)

# Step 2: Extract year
df['Year'] = df['Date of Payment'].dt.year

# Step 3: Calculate total per year
total_2024 = df[df['Year'] == 2024]['Amount'].sum()
total_2025 = df[df['Year'] == 2025]['Amount'].sum()

# Step 4: Calculate change
change = total_2025 - total_2024
change_pct = (change / total_2024 * 100).round(2)

# Step 5: Show results
print(f"ðŸ’° 2024 Total Spend: Â£{total_2024:,.2f}")
print(f"ðŸ’° 2025 Total Spend: Â£{total_2025:,.2f}")
print(f"ðŸ“Š Change Amount:    Â£{change:,.2f}")
print(f"ðŸ“Š Change %:         {change_pct}%")

if change > 0:
    print(f"\nâœ… Spending INCREASED by {change_pct}% in 2025")
else:
    print(f"\nâœ… Spending DECREASED by {abs(change_pct)}% in 2025")
```
